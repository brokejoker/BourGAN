```python
bourgan_config = {
    "epoch": 5000,
    "batch_size": 128,
    "alpha": 0.1,
    
    "scale_factor": 0.89,   #will not be used if using bourgain sampler
    
    "dataset":{
        "name": "gaussian_grid",
        "n": 5,
        "n_data": 50,
        "sig": 0.01
    },
    
    "use_gpu": True,
    "g_step": 1,
    "d_step": 1,
    "show_step": 500,
    
    "sampler":{
        "name": "bourgain",
        "path": None,
        "dist": "l2"
    },
    
#     "sampler":{
#         "name": "gaussian",
#         "dim": 25
#     },
    
    "nn_config_G":{
        "name": "DeepMLP_G",
        "hidden_size": 128
    },
    
    "nn_config_D":{
        "name": "DeepMLP_D",
        "hidden_size": 128
    },
    
    "opt_config_G":{
        "name": "adam",
        "default": False,
        "lr": 1e-3,
        "betas": (0.5, 0.999)
    },
    
    "opt_config_D":{
        "name": "adam",
        "default": False,
        "lr": 1e-3,
        "betas": (0.5, 0.999)
    },
    
    "zdist":{
        "name": "l2"
    },
    
    "gdist":{
        "name": "l2"
    }
}
```
